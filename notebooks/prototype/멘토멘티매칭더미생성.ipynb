```python
# 레벨 목록
levels = {
    1: "초급",
    2: "중급",
    3: "고급"
}
```

```python
def expand_keywords(keywords, synonyms_dict):
    expanded = set()
    for keyword in keywords:
        keyword_lower = keyword.lower()
        keyword_key = re.sub(r'\s+', '', keyword_lower)
        if keyword_key in synonyms_dict:
            expanded.update(synonyms_dict[keyword_key])
        else:
            expanded.add(keyword_key)
    return expanded
```

```python
def preprocess_keywords(keywords):
    processed = set()
    for keyword in keywords:
        keyword_clean = re.sub(r'\s+', '', keyword).lower()
        keyword_clean = re.sub(r'[^\w]', '', keyword_clean)
        processed.add(keyword_clean)
    return processed
```

```python
def determine_match(mentor, mentee):
    mentor_keywords = expand_keywords(mentor["expertise"], synonyms)
    mentee_keywords = expand_keywords(mentee["interests"], synonyms)
    mentor_keywords = preprocess_keywords(mentor_keywords)
    mentee_keywords = preprocess_keywords(mentee_keywords)
    common_keywords = mentor_keywords.intersection(mentee_keywords)
    level_difference = abs(mentor["level"] - mentee["level"])
    if common_keywords and level_difference <= 1:
        return 1  # 매칭 성공
    else:
        return 0  # 매칭 실패
```

```python
def generate_mentor(field):
    field_data = fields[field]
    name = random.choice(mentor_names)
    expertise_list = field_data["expertise"]
    max_k = min(len(expertise_list), 3)
    k = random.randint(1, max_k)
    expertise_sample = random.sample(expertise_list, k=k)
    level = random.choices([2, 3], weights=[1, 1], k=1)[0]
    bio_template = random.choice(field_data["mentor_bios"])
    bio = bio_template.format(expertise=", ".join(expertise_sample), level=levels[level])
    return {
        "name": name,
        "expertise": expertise_sample,
        "level": level,
        "bio": bio
    }
```

```python
fields_list = list(fields.keys())
data_list = []
num_pairs_per_field = 300  # 각 분야별로 300쌍 생성하여 총 4500쌍

for field in fields_list:
    for _ in range(num_pairs_per_field):
        mentor = generate_mentor(field)
        mentee = generate_mentee(field)
        match = determine_match(mentor, mentee)
        data_pair = {
            "field": field,
            "mentor": mentor,
            "mentee": mentee,
            "match": match
        }
        data_list.append(data_pair)

with open("mentor_mentee_data.json", "w", encoding="utf-8") as f:
    json.dump(data_list, f, ensure_ascii=False, indent=4)
```
