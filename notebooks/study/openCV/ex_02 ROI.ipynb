### ROI (Region of Image)
- 이미지 내의 내가 관심있는 영역 (이미지 내에 우리가 필요한 부분) </br>
ex) 전체 이미지에서 고양이가 있는 부분


```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/ironman.png', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# 출력할 이미지의 영역 (행(가로), 세로(열))
subimg = img[300:420, 20:230]


plt.xticks([]), plt.yticks([])
plt.imshow(img)

# (526, 481, 3)
```

### ROI에 도형 그리기

```python
img = cv2.imread('images/ironman.png', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 이미지에 선 그리기
# line(이미지, 시작점 좌표, 끝점 좌표, 색상, 선 두께)
img = cv2.line(img, (213, 20),(353, 250), (180, 230, 170), 3 )


plt.xticks([]), plt.yticks([])
plt.imshow(img)
# (526, 481, 3)
```

```python
img = cv2.imread('images/ironman.png', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 이미지에 네모 그리기
# rectangle(이미지, 좌상단 좌표, 우하단단 좌표, 색상, 선 두께)
img = cv2.rectangle(img, (200, 11),(353, 250), (180, 230, 170), 3)

plt.xticks([]), plt.yticks([])
plt.imshow(img)
# (526, 481, 3)
```

```python
img = cv2.imread('images/ironman.png', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 이미지에 원 그리기
# circle(이미지, 중점 좌표, 반지름, 색상, 선 두께) 선두께를 -1로 주면 내부가 채워진 도형이됨
img = cv2.circle(img, (280, 120), 115, (180, 230, 170), 3)

# 글자 표현
# 사용할 폰트 설정
font = cv2.FONT_HERSHEY_SIMPLEX
# putText(이미지, "텍스트", 좌표, 폰트, 크기,  색상, 선 굵기)
cv2.putText(img, "junsik", (240, 90), font, 1, (255,255,255), 3)

plt.xticks([]), plt.yticks([])
plt.imshow(img)
# (526, 481, 3)
```

### 한글 출력하기


```python
from PIL import ImageFont, ImageDraw, Image
import numpy as np
```

```python
img = cv2.imread('images/ironman.png', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 사용할 폰트와 크기 가져오기
font = ImageFont.truetype("fonts/gulim.ttc", 30)

# np의 array 이미지를 PIL 이미지로 변환
img_pil = Image.fromarray(img)

# PIL 이미지를 그림
draw = ImageDraw.Draw(img_pil)
# PIL 이미지에 텍스트 그리기 text( (좌표(위치)), "텍스트", font=font, 색상)
draw.text( (230, 70), "엄준식", font=font, fill=(255, 255 , 255))

# PIL 이미지를 numpy 배열로 변환
img = np.array(img_pil)

plt.xticks([]), plt.yticks([])
plt.imshow(img)
# (526, 481, 3)
```
