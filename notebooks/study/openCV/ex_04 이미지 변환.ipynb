#### 이미지 이동

```python
img = cv2.imread('images/shinchangsub.png', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

h, w, channel = img.shape

# 이동 행렬 만들기 (20, 20 만큼 이동하라)
M = np.float32([[1, 0, -20],
            [0, 1, -20]])

# 행렬(M)에 따른 변환을해주는 함수
result = cv2.warpAffine(img, M, (w, h), -1, cv2.INTER_LINEAR, cv2.BORDER_REFLECT)
# interpolation : 보간법 (이미지 변환 시 깨지는 부분을 보간, 일부 정보만을 가지고 빈 부분을 복원하는 방법)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.xticks([]), plt.yticks([])
plt.imshow(img)
plt.title("Original")

plt.subplot(1, 2, 2)
plt.xticks([]), plt.yticks([])
plt.imshow(result)
plt.title("Result")
```

### 이미지 회전

```python
img = cv2.imread('images/shinchangsub.png', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

h, w, channel = img.shape

# 회전 각 설정 (회전중심점, 회전각, 확대 비율)
M = cv2.getRotationMatrix2D((w/2, h/2), 20, 1) # 회전 행렬을 만들어주는 함수

# 행렬(M)에 따른 변환을해주는 함수
result = cv2.warpAffine(img, M, (w, h), -1, cv2.INTER_LINEAR, cv2.BORDER_REFLECT)
# interpolation : 보간법 (이미지 변환 시 깨지는 부분을 보간, 일부 정보만을 가지고 빈 부분을 복원하는 방법)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.xticks([]), plt.yticks([])
plt.imshow(img)
plt.title("Original")

plt.subplot(1, 2, 2)
plt.xticks([]), plt.yticks([])
plt.imshow(result)
plt.title("Result")
```

### 이미지 기울임

```python
img = cv2.imread('images/shinchangsub.png', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

h, w, channel = img.shape

ori = np.float32([[0,0], [w, 0], [0, h]])
trans = np.float32([[0,0], [w-100, 0], [100, h]])

# 3 좌표를 이용해 변환행렬 계산
M = cv2.getAffineTransform(ori, trans)

result = cv2.warpAffine(img, M, (w, h))

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.xticks([]), plt.yticks([])
plt.imshow(img)
plt.title("Original")

plt.subplot(1, 2, 2)
plt.xticks([]), plt.yticks([])
plt.imshow(result)
plt.title("Result")
```

#### 원근 변환
- Affine (이미지 기울이기)와는 달리 4개 점을 이용하여 이미지 변환

```python
# 이미지 출력력

img = cv2.imread('images/presentation.png', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

h, w, channel = img.shape

# 왼쪽위, 오른쪽위, 왼쪽 아래, 오른쪽 아래
ori_img = np.float32([[98, 49], [499, 96], [18, 705], [572, 695]]) # 원래 이미지 좌표 (그림판으로 갖다대면서 확인)
trans_img = np.float32([[0, 0], [w, 0], [0, h], [w, h]])

# 행렬 구하기
M = cv2.getPerspectiveTransform(ori_img, trans_img)

result = cv2.warpPerspective(img, M, (w, h))

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.xticks([]), plt.yticks([])
plt.imshow(img)
plt.title("Original")

plt.subplot(1, 2, 2)
plt.xticks([]), plt.yticks([])
plt.imshow(result)
plt.title("Result")
```
